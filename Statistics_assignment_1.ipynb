```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=np.array([7,9,9,10,10,10,10,11,12,12,14,14])
data.sort()
size=data.size
size
if(size%2==1):
    mid=int((size-1)/2)
    if mid%2==1:
        
        Q1=data[int((mid-1)/2)]
        Q2=data[int((size-(mid-1)/2)-1)]
        
    else:
        
        Q1=(data[int((mid-1)/2)]+data[int(((mid-1)/2)-1)])/2
        Q2=(data[int((size-(mid-1)/2)-1)]+data[int((size-(mid-1)/2)-2)])/2
else:

    if (size/2)%2==1:
    
        mid=int((size-1)/2)
        Q1=data[int((mid-1)/2)]
        Q2=data[int((size-(mid-1)/2)-1)]
    
    else:
    
        Q1=(data[int((mid-1)/2)]+data[int(((mid-1)/2)-1)])/2
        Q2=(data[int((size-(mid-1)/2)-1)]+data[int((size-(mid-1)/2)-2)])/2
    

IQR=Q2-Q1
IQR
```
